```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pybaseball as pyb
```

```python
# find justin verlander's player id
lw = pyb.playerid_lookup('webb', 'logan')
lw
```

```python
lw_id = lw['key_mlbam'][0]
```

```python
start_date = '2019-01-01'
end_date = '2023-12-31'
```

```python
lw_stats = pyb.statcast_pitcher(start_date, end_date, lw_id)
lw_stats
```

```python
# convert game_date to datetime
lw_stats['game_date'] = pd.to_datetime(lw_stats['game_date'])
```

```python
# drop deprecated columns and unneeded columns
deprecated_cols = ['spin_dir', 'spin_rate_deprecated', 'break_angle_deprecated', 'break_length_deprecated', 'tfs_deprecated', 'tfs_zulu_deprecated', 'umpire']
unneeded_cols = ['pitcher', 'pitcher.1', 'fielder_2.1', 'fielder_3', 'fielder_4', 'fielder_6', 'fielder_7', 'fielder_8', 'fielder_9']
lw_stats = lw_stats.drop(columns= deprecated_cols + unneeded_cols)
```

```python
# remove rows where pitch_type is NaN
lw_stats = lw_stats.dropna(subset=['pitch_type']).reset_index(drop=True)
lw_stats
```

```python
### on_1b
# if not NaN, replace with 1
# if NaN, replace with 0
lw_stats['on_1b'] = lw_stats['on_1b'].notnull().astype(int)

### on_2b
# if not NaN, replace with 1
# if NaN, replace with 0
lw_stats['on_2b'] = lw_stats['on_2b'].notnull().astype(int)

### on_3b
# if not NaN, replace with 1
# if NaN, replace with 0
lw_stats['on_3b'] = lw_stats['on_3b'].notnull().astype(int)
```

```python
# score difference
lw_stats['score_diff'] = lw_stats['fld_score'] - lw_stats['bat_score']

# create a new column: score_lead
# if score_diff is positive, replace with 1
# if score_diff is negative, replace with -1
# if score_diff is 0, replace with 0
lw_stats['score_lead'] = np.sign(lw_stats['score_diff'])
```

```python
def create_previous_pitch_type(df):
    prev_pitch_type = []

    for idx, row in df.iterrows():
        # if first row
        if (idx == 0):
            if (row['inning'] == 1) and (row['pitch_number'] == 1):
                prev_pitch_type.append(np.nan)
                continue
            else:
                prev_pitch_type.append(np.inf)
                continue

        # if first pitch of plate appearance, append NaN
        if row['pitch_number'] == 1:
            prev_pitch_type.append(np.nan)
        else:
            # else get the pitch type of the previous pitch
            # check for same plate appearance
            if (df.iloc[idx-1]['game_date'] == row['game_date']) and (df.iloc[idx-1]['at_bat_number'] == row['at_bat_number']):
                prev_pitch_type.append(df.iloc[idx-1]['pitch_type'])
            else:
                prev_pitch_type.append(np.inf)
    
    return prev_pitch_type
    
```

```python
def create_previous_description(df):
    prev_description = []

    for idx, row in df.iterrows():
        # if first row
        if (idx == 0):
            if (row['inning'] == 1) and (row['pitch_number'] == 1):
                prev_description.append(np.nan)
                continue
            else:
                prev_description.append(np.inf)
                continue

        # if first pitch of first inning, append NaN
        if (df.iloc[idx-1]['game_date'] != row['game_date']) and (row['pitch_number'] == 1) and (row['inning'] == 1) and (row['at_bat_number'] == df[df['game_date'] == row['game_date']]['at_bat_number'].min()):
            prev_description.append(np.nan)
            continue

        # check for same game date
        if (df.iloc[idx-1]['game_date'] == row['game_date']):
            # check for same plate appearance
            if (df.iloc[idx-1]['at_bat_number'] == row['at_bat_number']):
                # check for correct count
                if ((df.iloc[idx-1]['balls'] + 1 == row['balls']) or (df.iloc[idx-1]['strikes'] + 1 == row['strikes'])) or ((df.iloc[idx-1]['balls'] == row['balls']) and (df.iloc[idx-1]['strikes'] == row['strikes'])):
                    prev_description.append(df.iloc[idx-1]['description'])
                else:
                    prev_description.append(np.inf)
            # new plate appearance, same inning
            elif (df.iloc[idx-1]['inning'] == row['inning']) and (df.iloc[idx-1]['at_bat_number'] + 1 == row['at_bat_number']):
                prev_description.append(df.iloc[idx-1]['description'])
            # new inning
            elif (df.iloc[idx-1]['inning'] + 1 == row['inning']) and (row['pitch_number'] == 1):
                prev_description.append(df.iloc[idx-1]['description'])
            else:
                prev_description.append(np.inf)
        else:
            prev_description.append(np.inf)
        

    return prev_description
```

```python
def create_new_columns(df):
    # create new columns
    df['prev_pitch_type'] = create_previous_pitch_type(df)
    df['prev_description'] = create_previous_description(df)

    # remove rows where prev_pitch_type or prev_description is inf
    df = df[(df['prev_pitch_type'] != np.inf) & (df['prev_description'] != np.inf)]
    
    # reset index 
    df = df.reset_index(drop=True)
    
    return df
```

```python
# reverse the dataframe
lw_stats = lw_stats.iloc[::-1].reset_index(drop=True)
```

```python
# add new columns
lw_stats_modified = create_new_columns(lw_stats)
lw_stats_modified
```

```python
# export to csv
lw_stats_modified.to_csv('logan_webb_stats.csv', index=True)
```
