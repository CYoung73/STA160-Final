```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import pybaseball as pyb
import seaborn as sns
from scipy import stats
from sklearn.cluster import AgglomerativeClustering
```

```python
pyb.cache.enable()
```

```python
data_19 = pyb.statcast(start_dt="2019-01-01", end_dt="2019-12-31")
```

```python
data_21 = pyb.statcast(start_dt="2021-01-01", end_dt="2021-12-31")
data_22 = pyb.statcast(start_dt="2022-01-01", end_dt="2022-12-31")
data_23 = pyb.statcast(start_dt="2023-01-01", end_dt="2023-12-31")
```

```python
data = data_19.append(data_21)
data = data.append(data_22)
data = data.append(data_23)
```

```python
filtered_data = data[data['player_name'] == 'Webb, Logan']
```

```python
filtered_data.columns
```

```python
p_2019 = filtered_data.loc[filtered_data['game_date'].dt.year == 2019]
p_2021= filtered_data.loc[filtered_data['game_date'].dt.year == 2021]
p_2022= filtered_data.loc[filtered_data['game_date'].dt.year == 2022]
p_2023= filtered_data.loc[filtered_data['game_date'].dt.year == 2023]
```

```python
launch_data = filtered_data[['game_date','pitch_type','launch_angle', 'launch_speed']].replace({pd.NA: np.nan}).dropna()
launch_data_19 = p_2019[['game_date','pitch_type','launch_angle', 'launch_speed']].replace({pd.NA: np.nan}).dropna()
launch_data_21 = p_2021[['game_date','pitch_type','launch_angle', 'launch_speed']].replace({pd.NA: np.nan}).dropna()
launch_data_22 = p_2022[['game_date','pitch_type','launch_angle', 'launch_speed']].replace({pd.NA: np.nan}).dropna()
launch_data_23 = p_2023[['game_date','pitch_type','launch_angle', 'launch_speed']].replace({pd.NA: np.nan}).dropna()
```

```python
# Launch Speed Over Time 

grouped = launch_data.groupby('game_date')['launch_speed'].agg(['mean', 'count', 'std'])

plt.figure(figsize=(14, 7))
plt.scatter(launch_data['game_date'], launch_data['launch_speed'], alpha=0.3)
plt.plot(grouped.index, grouped['mean'], color='red', linestyle='--', label='Mean Launch Speed')
plt.xlabel('Date')
plt.ylabel('Launch Speed')
plt.title('Launch Speed Over Time')
plt.legend()
```

```python
# Function that plots Launch Speed Over Time
def plot_LS(ax,dt,title):
    grouped = dt.groupby('game_date')['launch_speed'].agg(['mean', 'count', 'std'])
    ax.scatter(dt['game_date'], dt['launch_speed'], alpha=0.3)
    ax.plot(grouped.index, grouped['mean'], color='red', linestyle='--', label='Mean Launch Speed')
    ax.set_xlabel('Date')
    ax.set_ylabel('Launch Speed')
    ax.set_title(title)
    ax.legend()
```

```python
# Launch Speed Over Time by Year
fig, axs = plt.subplots(2, 3, figsize=(14, 10))
plot_LS(axs[0, 0], launch_data_19, 'Launch Speed in 2019')
plot_LS(axs[0, 1], launch_data_21, 'Launch Speed in 2021')
plot_LS(axs[0, 2], launch_data_22, 'Launch Speed in 2022')
plot_LS(axs[1, 0], launch_data_23, 'Launch Speed in 2023')
axs[1,2].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# Launch Angle Over Time 

grouped = launch_data.groupby('game_date')['launch_angle'].agg(['mean', 'count', 'std'])

plt.figure(figsize=(14, 7))
plt.scatter(launch_data['game_date'], launch_data['launch_angle'], alpha=0.3, label='Individual Data Points')
plt.plot(grouped.index, grouped['mean'], color='red', linestyle='--', label='Mean Launch Speed')
plt.xlabel('Date')
plt.ylabel('Launch Angle')
plt.title('Launch Angle Over Time')
plt.legend()
```

```python
# Launch Angle Over Time by Year
fig, axs = plt.subplots(2, 3, figsize=(14, 10))
plot_LA(axs[0, 0], launch_data_19, 'Launch Angle in 2019')
plot_LA(axs[0, 1], launch_data_21, 'Launch Angle in 2021')
plot_LA(axs[0, 2], launch_data_22, 'Launch Angle in 2022')
plot_LA(axs[1, 0], launch_data_23, 'Launch Angle in 2023')
axs[1,2].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
def cluster_furthest(data, n_clusters=2):
    model = AgglomerativeClustering(n_clusters=n_clusters, linkage='complete')
    model.fit(data)
    labels = model.labels_
    return labels
```

```python
cluster_labels =cluster_furthest(launch_data_21[['launch_angle','launch_speed']])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(launch_data_21['launch_angle'], launch_data_21['launch_speed'], c=cluster_labels, cmap='viridis', alpha=0.8)
plt.title('Scatter Plot with Cluster Labels')
plt.xlabel('Launch Angle')
plt.ylabel('Launch Speed')
plt.legend()
plt.show()
```
